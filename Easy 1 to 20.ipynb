```python
def is_prime(n):
    if n == 0 or n == 1:
        return f"The number {n} is not prime."
    for i in range(2, n):
        if n % i == 0:
            return f"The number {n} is not prime."
    return f"The number {n} is prime."

is_prime(8)
```

```python
def sec_lar(lst):
    lst.sort()
    return lst[-2]

sec_lar([2, 5, 10, 23, 24, 10])
```

```python
def swap(a, b):
    a, b = b, a
    return a, b

swap(3, 6)
```

```python
def rev_str(str):
    rev = ''
    for i in range(len(str) - 1, -1, -1):
        rev += str[i]
    return rev

rev_str('python')
```

```python
def pali(str):
    if str == str[::-1]:
        return f"The string '{str}' is a palindrome."
    else:
        return f"The string '{str}' is not a palindrome."

pali('malayalam')
```

```python
def fact_nor(n):
    ans = 1
    for i in range(1, n + 1):
        ans *= i
    return ans

def fact_rec(n):
    if n == 1:
        return n
    return n * fact_rec(n - 1)

print(fact_nor(5))
print(fact_rec(5))
```

```python
def max_min(lst):
    max_ = min_ = lst[0]
    for num in lst:
        if num > max_:
            max_ = num
        if num < min_:
            min_ = num
    return(max_, min_)

max_min([1, 2, 3, 10, 0, -1])
```

```python
def anag(str1, str2):
    if len(str1) != len(str2):
        return "Not anagrams."
    for ele in str1:
        if ele not in str2:
            return "Not anagrams."
    return "Anagrams!"

anag("asif", "fisa")
```

```python
def sum_dig(n):
    ans = 0
    while n != 0:
        ans += n % 10
        n //= 10
    return ans

sum_dig(12349)
```

```python
def odd_even(lst):
    e = []
    o = []
    for ele in lst:
        if ele % 2 == 0:
            e.append(ele)
        else:
            o.append(ele)
    return e, o

odd_even([1, 2, 3, 4, 5, 6])
```

```python
def dupe(lst):
    ans = []
    unique = list(set(lst))
    for x in unique:
        c = 0
        for y in lst:
            if x == y:
                c += 1
        if c > 1:
            ans.append(x)
    return ans

dupe([1, 1, 'asif', 'lemon', 1, 'lemon', 'man', 'man'])
```

```python
def miss(lst):
    for i in range(len(lst) - 1):
        if lst[i + 1] != lst[i] + 1:
            return lst[i] + 1

miss([129, 130, 132])
```

```python
def freq(lst):
    dict_ = {}
    for ele in lst:
        dict_[ele] = dict_.get(ele, 0) + 1
    return dict_

freq(['1', '2', '2', '3', '2', 2, 1, 1])
```

```python
def merge(lst1, lst2):
    i = j = 0
    fin = []
    while i < len(lst1) and j < len(lst2):
        if lst1[i] < lst2[j]:
            fin.append(lst1[i])
            i += 1
        else:
            fin.append(lst2[j])
            j += 1
    fin.extend(lst1[i:])
    fin.extend(lst2[j:]) 
    return fin

merge([1, 2, 10, 30], [10, 15, 20])
```

```python
def check_sort(lst):
    for i in range(len(lst) - 1):
        if lst[i] > lst[i + 1]:
            return "Not sorted!"
    return "Sorted!"

check_sort([1, 2, 3, 2, 4]) 
```

```python
def inter(lst1, lst2):
    fin = []
    if len(lst1) < len(lst2):
        for ele in lst1:
            if ele in lst2:
                fin.append(ele)
    else:
        for ele in lst2:
            if ele in lst1:
                fin.append(ele)
    return fin

inter([1, 2, 4, 10, 100], [4, 5, 100])
```

```python
def str_oc(str, char):
    c = 0
    for ele in str:
        if ele == char:
            c += 1
    return c

str_oc('supercalifragilisticexpialidocious', 's')
```

```python
def rev_sent(sent):
    lst = sent.split()
    lst = lst[::-1]
    fin = ''
    for ele in lst:
        fin += ele
        fin += ' '
    return fin[:-1]

rev_sent("Asif is trying to get a job")
```

```python
def fib(n):
    a = 0
    b = 1
    fin = [a, b]
    for i in range(n - 2):
        c = a + b
        fin.append(c)
        a = b
        b = c
    return fin

fib(10)
```
